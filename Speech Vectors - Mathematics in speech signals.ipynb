# Vectors:
<br>
<p>In mathematics, a vector is a mathematical object that represents a quantity characterized by both magnitude and direction. Vectors are commonly used to represent physical quantities such as displacement, velocity, acceleration, force, and more.</p>
<p>In linear algebra, Vectors are typically represented as collection of numbers in an array, where each elements corresponds to component of array in a perticular direction.</p>
<p>By treating speech signals as vectors we can understand how these signals can be manipulated, transformed, and analyzed using principles from vector algebra.</p>

```python
# !pip install sounddevice
import sounddevice as sd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# "sounddevice"
<br>
<li>"sounddevice" python library is a user friendly package to record/play audio signals.</li>
<li>It provides an easy interface to interact with the sound devices.</li>
<li>It can be seemlessly integrated with numpy to facilitate various analytical applications</li>

```python
# Parameters to record audio using sounddevice library:
# frames, samplerate and channels are the basic parameters to be passed in this context.
# "Frames" is product of sample rate and duration(lenght of the audio array).
# "samplerate" is the number of audio sample recorded per second (For CD quality: 8000)
# "channels" is the number of audio channels to record. (For monarual: channels = 1 | for stereo: channels = 2)

duration=30
fs=8000  #44000Hhz
my_record=sd.rec(fs*duration,samplerate=fs,channels=1)
print('speak')
sd.wait()
```

```python
sd.play(my_record,fs)
print("Please note that I spoke the numbers 1 to 10 within the 30 seconds of recording time")
sd.wait()
```

<h3>Speech Array</h3>

<p>Since the recorded sound is of single channel the audio is stored in form of 1D array.</p>

```python
my_record.shape
# 8000 samples/sec x 30 Sec 
```

<h3>Visualizing audio sample</h3>

<p>Audio signals when plotted shows deflections, this can be used to infer the number of words spoken in the given frame.</p> 

```python
plt.figure(figsize=(10,2),dpi=150)
plt.plot(my_record)
plt.show()
```

<h3>Basic array operations on the audio sample</h3>

<li>Array slicing: Slicing the first 15 seconds of the audio. </li>

```python
my_record[:(my_record.shape[0])//2]
```

<li> Down sampling the vector by 2.</li>

```python
down_sampled = my_record[0:(my_record.shape[0]):2]
sd.play(down_sampled,fs)
```

<li>Upsampling the vector by 2</li>

```python
upsampled_array = np.repeat(upsampled_array, 1)
sd.play(upsampled_array,fs)
```

<h3>Vector Transformations</h3>
<p>Applying transformations on audio arrays enables us to extract meaningful information, enhance audio quality, manipulate audio signals for creative purposes, and develop advanced audio processing algorithms and applications.</p>

 <li> 3*Speech signal </li>
 <p>Note that the volume of the sound is spiked.</p>

```python
sd.play(3*(my_record),fs)
plt.figure(figsize=(10,2),dpi=150)
plt.plot(3*(my_record))
plt.show()
```

 <li> 0.5*Speech signal </li>
 <p>Note that the volume of the sound is reduced.</p>

```python
sd.play(0.5*(my_record),fs)
plt.figure(figsize=(10,2),dpi=150)
plt.plot(0.5*(my_record))
plt.show()
```

 <li>exp(0.5*Speech signal) </li>
 <p>Note that the sound is slightly distorted.</p>

```python
sd.play(np.exp(0.5*my_record),fs)
plt.figure(figsize=(10,2),dpi=150)
plt.plot(np.exp(0.5*my_record))
plt.show()
```

<li>ln(speech signal)</li>
<p>Note that the sound is inaudible but the change can be visualized the plot.</p>

```python
sd.play(np.log(my_record),fs)
plt.figure(figsize=(10,2),dpi=150)
plt.plot(np.log(my_record))
plt.show()
```

<li>Cos (Speech Signal)</li>
<p>Note that the sound is distorted and the volume is low.</p>

```python
sd.play(np.cos(my_record),fs)
plt.figure(figsize=(10,2),dpi=150)
plt.plot(np.cos(my_record))
plt.show()
```

<h3>Interesting result of vector addition</h3>
<p>Let us record noise sample of same duration and add it to our main recording.(Record you favorite song) </p>

```python
duration=30
fs=8000  #44000Hhz
my_noise=sd.rec(fs*duration,samplerate=fs,channels=1)
print('speak')
sd.wait()
```

```python
sd.play(my_noise+my_record,fs)
plt.figure(figsize=(10,2),dpi=150)
plt.plot(my_noise+my_record)
plt.show()
```

<p>Audio signals can be amalgamated, affording the opportunity to execute analogous transformations for processing audio files.</p>
